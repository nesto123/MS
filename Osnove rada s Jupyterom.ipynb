## Primjeri

### Osnove rada s Jupyterom

Tekst se nalazi u ćelijama, ima ih tri vrste:
* ulazne ćelije, gdje pišemo kod u Pythonu, označene s `In[`*broj*`]`
* izlazne ćelije, gdje Python javlja rezultate, označene s `Out[`*broj*`]`
* tekstne ćelije, gdje pišemo tekst u Markdownu

*broj*evi gore se odnose na vremenski tijek evaluacije pojedinih ćelija, koji se ne mora podudarati s prostornim. Ipak, za uredne dokumente, dobro je da se bilježnica može evaluirati "odozgo prema dolje" (Menu `Kernel`... `Restart & Run All`). U bilo kojem trenutku možemo s `_`*broj* referirati na rezultat pojedine naredbe (skraćeno samo `_` za zadnju izvršenu naredbu), iako je naravno bolje rezultatima dati lijepa imena ako znamo da ćemo ih još koristiti.

Trenutno odabranu ćeliju možemo pretvoriti u Python pomoću `Esc` `Y` ili u Markdown pomoću `Esc` `M`. Općenito, `Esc` nas prebacuje u način rada s ćelijama (plavi obrub) gdje pored promjene tipa određene ćelije možemo zadavati razne druge naredbe (`Help`... `Keyboard shortcuts` za popis svih, ovdje su najčešće korištene):

* `C`, `X`, `V`: kopiranje, rezanje i lijepljenje ćelija
* `J`, `K` ili strelice dolje/gore: kretanje između ćelija (+`Shift`: odabir više ćelija)
* `A`, `B`: dodavanje nove ćelije iznad/ispod prethodne
* `DD`: brisanje trenutno odabrane ćelije (dvostruki pritisak na `D`)

Klik u ulaznu ćeliju (ili `Enter` kada je odabrana) pokreće način rada s kodom u ćeliji (zeleni obrub) gdje možemo pisati kod. Izvršavamo ga sa `Shift`+`Enter`. Također, dvostruki klik u tekstnu ćeliju (ili `Enter`) otvara nam _source_ (Markdown) te ćelije, gdje možemo vidjeti kako su pojedini efekti postignuti. `Shift`+`Enter` renderira tekstnu ćeliju kao običan (formatirani) tekst. Trostruki klik u izlaznu ćeliju odabire čitav izlaz u ćeliji kako bismo ga mogli prenijeti drugamo.

### Kompleksniji primjer

Učitamo potrebne biblioteke: [`skimage`](https://scikit-image.org/docs/dev/api/skimage.html) za rad sa slikama i [`matplotlib`](https://matplotlib.org/) za vizualizaciju. Dohvatimo sliku novčićâ iz biblioteke primjera i prikažemo je kao crno-bijelu.

```python
from skimage import data
import matplotlib.pyplot as plt
coins = data.coins()
plt.imshow(coins, cmap='gray')
```

`.shape` je uređeni par koji predstavlja dimenzije slike: `(`*redci*`, `*stupci*`)`. Operatorom `[]` možemo iz slike "vaditi" pojedine piksele ili isječke: ovdje smo izrezali novčić koji je prikazan od 10. do 80. retka i od 300. do 370. stupca.

```python
coins.shape
```

```python
coins_zoom = coins[10:80, 300:370]
```

Veliki paketi često sadrže podpakete koji se učitavaju isto kao i objekti/klase/funkcije. `skimage.restoration` je zbirka filtera koje možemo primjenjivati na slike. Jedan od njih je [`denoise_tv_chambolle`](https://scikit-image.org/docs/dev/api/skimage.restoration.html#skimage.restoration.denoise_tv_chambolle).

```python
from skimage import restoration
tv_coins = restoration.denoise_tv_chambolle(coins_zoom, weight=.05)
```

Prikazujemo isječke jedan pored drugog: lijevo prije primjene filtera, desno poslije.

```python
slika, (lijevo, desno) = plt.subplots(ncols=2)
lijevo.imshow(coins_zoom)
desno.imshow(tv_coins)
```

### Interaktivnost

Jupyter ima brojne [mogućnosti za interaktivni rad](https://ipywidgets.readthedocs.io/en/latest/examples/Using%20Interact.html). Pokažimo samo neke.

```python
from ipywidgets import interact
interact(lambda x=8, y=1, brže=True: x + y**2 if brže else x + 2*y)
```

```python
import networkx as nx
interact(lambda n=(2, 30), m=(1, 10), p=(0, 1, 1e-3), generator={
                   'lobster': lambda n, m, p: nx.random_lobster(n, p, p / m),
                 'power law': lambda n, m, p: nx.powerlaw_cluster_graph(n, m, p),
     'Newman-Watts-Strogatz': lambda n, m, p: nx.newman_watts_strogatz_graph(n, m, p),
               'Erdős-Rényi': lambda n, m, p: nx.erdos_renyi_graph(n, p),
}: nx.draw(generator(n, m, p)));
```
